```python
import pandas as pd

order["Order Date"] = pd.to_datetime(
    order["Order Date"],
    unit="D",
    origin="1899-12-30"
)

order["Ship Date"] = pd.to_datetime(
    order["Ship Date"],
    unit="D",
    origin="1899-12-30"
)
```

```python
quantity_map = {
    "one":1,
    "two":2,
    "three":3,
    "four":4,
    "five":5,
    "six":6,
    "seven":7,
    "eight":8,
    "nine":9,
    "ten":10
}

order["Quantity"]=(
    order["Quantity"].replace(quantity_map).astype(int)
)
```

```python
sorted(order["Quantity"].unique())
```

```python
order["Order Year"] = order["Order Date"].dt.year

order.groupby("Order Year").agg({
    "Sales": "sum",
    "Profit": "sum",
    "Order ID": "nunique"
})
```

```python
yearly = order.groupby("Order Year").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Order ID":"nunique"
})

yearly["Sales YoY %"] = yearly["Sales"].pct_change()*100
yearly["Sales YoY %"] = yearly["Profit"].pct_change()*100
yearly["Sales YoY %"] = yearly["Order ID"].pct_change()*100

yearly.round(2)
```

```python
order.groupby("Category").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Order ID":"nunique"
}).sort_values("Sales", ascending=False)
```

```python
order.groupby("Category").agg({
    "Sales":"sum",
    "Profit":"sum"
}).assign(
    Profit_Margin=lambda x: x["Profit"] / x["Sales"] * 100
).round(2)
```

```python
order.groupby("Region").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Order ID":"nunique"
}).sort_values("Sales", ascending=False)
```

```python
region = order.groupby("Region").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Order ID":"nunique"
})

region["Sales Share %"] = (
    region["Sales"] /
    region["Sales"].sum() * 100
)

region.round(2)
```
